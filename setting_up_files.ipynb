```python
import os
import shutil 
# Remove .git files so git tracks the whole project together instead of having 
# two separate git entities. 

def remove_git(directory):

    for root, dirs, files in os.walk(directory):
        for dir in dirs:
        
                if os.path.basename(dir) == '.git': 
                    print('Found .git directory at.....', root + "/" + dir)
                    shutil.rmtree(root + "/" + dir)
                    print(dir, "file has been located and removed")
                else: 
                    dotgit=False
    
    for root, dirs, files in os.walk(directory):            
        for file in files:         
                if os.path.basename(file) == '.gitignore':
                    print('Found .gitignore directory at.....', root + "/" + file)
                    os.remove(root + "/" + file)
                    print(file, "file has been located and removed")
                else: 
                    gitignore=False

    if not gitignore:
        print("No gitignores in," , directory )
    if not dotgit:
        print("No .git directories here either")
    print("------------------")


remove_git('/content/drive/My Drive/BiasMitigation/cityscapesScripts')
remove_git('/content/drive/My Drive/BiasMitigation/learning-not-to-learn')
```

```python
def ipynb_to_py(directory):

    import os
    ! pip install nbconvert

    for subdir, dirs, files in os.walk(directory):
        for file in files:
            if os.path.splitext(file)[1] == '.ipynb':
                exe_command = 'jupyter nbconvert "{}" --to python'.format(''.join(os.path.join(subdir, file)))
                os.system(exe_command)
                
                # Delete the file if you dont want duplicates remove hash below
                # os.remove(''.join(os.path.join(subdir, file)))
            else:
                continue
        print("Conversion Complete!")
```
