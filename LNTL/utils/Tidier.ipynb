```python
# File Tidier:

import os 
import shutil 

''' We dont need to keep all the checkpoint files for each model run 
    only the best model is needed to resume training from this checkpoint
    this script selects the last checkpoint file in the checkpoints 
    folder and deletes the rest. '''


def tidier(network_name, model_id):
    
    project_dir = "/content/drive/MyDrive/" + network_name
    checkpoints = project_dir + "training_logs/model_" + model_id + "/checkpoints"

    last = os.listdir(checkpoints)[-1]
    folder = os.listdir(checkpoints)

    for filename in folder:
        if filename != last:
        
            os.remove(checkpoints + "/" + filename)
            #print(checkpoints + "/" + filename)
        
    print("File structure has been tidied up, only the last checkpoint file remains")
```
